```python
df["BloodPressure"] = df["BloodPressure"].replace(0,df["BloodPressure"].median())
```

```python
df["BMI"] = df["BMI"].replace(0,df["BMI"].median())
df["Insulin"] = df["Insulin"].replace(0,df["Insulin"].median())
df["SkinThickness"] = df["SkinThickness"].replace(0,df["SkinThickness"].median())
```

```python
fig,axes = plt.subplots(3,2,figsize = (16,10))
sns.histplot(df["Glucose"],ax=axes[0,0],kde=True)
sns.histplot(df["BMI"],ax=axes[0,1],kde=True)
sns.histplot(df["BloodPressure"],ax=axes[1,0],kde=True)
sns.histplot(df["Insulin"],ax=axes[1,1],kde=True)
sns.histplot(df["SkinThickness"],ax=axes[2,0],kde=True)
```

```python
plt.figure(figsize=(13,8))
sns.boxplot(df)
plt.xticks(rotation=45)
```

```python
## Segregate with dependent and independent feature
X = df.drop(columns=["Outcome"])
y = df["Outcome"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =  train_test_split (X,y,test_size=0.25,random_state=0)
```

```python
import pickle
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale =scaler.transform(X_test)

file = open("Model/standardscaler.pkl","wb")
pickle.dump(scaler,file)
```

```python
plt.figure(figsize = (13,8))
sns.boxplot(X_train_scale)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train_scale,y_train)
```

```python
y_pred = log_reg.predict(X_test_scale)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
## Hyperparameter tuning
from sklearn.model_selection import GridSearchCV

parameter = {"penalty" : ['l1', 'l2', 'elasticnet'],
            'C' :np.logspace(-3,3,7),
            'solver':["liblinear","lbfgs","newton-cg"]
            }
```

```python
logreg = LogisticRegression()
clf = GridSearchCV(logreg,
            parameter,
            scoring="accuracy",
            cv=10)
clf.fit(X_train_scale,y_train)
```

```python
clf.best_params_
```

```python
classifier = LogisticRegression(C=1.0,penalty = "l2",solver = "liblinear")
```

```python
classifier.fit(X_train_scale,y_train)
```

```python
y_pred = classifier.predict(X_test_scale)
```

```python
y_pred
```

```python
print(confusion_matrix(y_test,classifier.predict(X_test_scale)))
```

```python
print(accuracy_score(y_test,classifier.predict(X_test_scale)))
```

```python
print(classification_report(y_test,classifier.predict(X_test_scale)))
```

```python
file = open("Model/classifier.pkl","wb")
pickle.dump(classifier,file)
file.close()
```
